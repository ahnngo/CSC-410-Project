```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Clean Data.csv')
```

```python
df.head()
```

```python
target = []
for hin, schi, citi, bio in zip(df['Hinselmann'], df['Schiller'], df['Citology'], df['Biopsy']):
    if hin == 1 or schi == 1 or citi == 1 or bio == 1:
        target.append(1)
    else:
        target.append(0)
```

```python
target.count(1)
```

```python
df['Target'] = target
```
